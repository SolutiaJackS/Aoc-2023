```python
import requests

url = "https://adventofcode.com/2023/day/5/input"

headers = {
    'Cookie': 'session={your session}'
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    # Print the content of the response
    pass
else:
    # Print an error message if the request was not successful
    print(f"Error: {response.status_code}")
    print(response.text)

lines = response.text.split('\n')[:-1]
```

```python
def get_value_by_map(seed, mapping):
    soil_value = None
    for entry in mapping:
        if seed >= entry['input'] and seed < entry['input'] + entry['range']:
            seed_dif = seed - entry['input']
            soil_value = entry['output'] + seed_dif
    if soil_value is None:
        soil_value = seed
    return soil_value

locations = []
for seed in seeds:
    soil = get_value_by_map(int(seed), seed_to_soil_dict)
    fertilizer = get_value_by_map(soil, soil_to_fertilizer_dict)
    water = get_value_by_map(fertilizer, fertilizer_to_water_dict)
    light = get_value_by_map(water, water_to_light_dict)
    temperature = get_value_by_map(light, light_to_temperature_dict)
    humidity = get_value_by_map(temperature, temperature_to_humidity_dict)
    location = get_value_by_map(humidity, humidity_to_location_dict)
    locations.append(location)

print(locations)
min(locations)
```
